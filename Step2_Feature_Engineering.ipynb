```python
import pandas as pd
import numpy as np
```

```python
# 1. Chargement des données nettoyées à l'étape 1
data = pd.read_csv('data_step1.csv')
```

```python
# 2. Création de la colonne 'comb'
# C'est la "soupe de mots" qui va permettre au moteur de comparer les films
data['comb'] = data['actor_1_name'] + ' ' + data['actor_2_name'] + ' ' + data['actor_3_name'] + ' ' + data['director_name'] + ' ' + data['genres']
```

```python
# 3. Suppression des doublons si certains titres apparaissent plusieurs fois
data.drop_duplicates(subset="movie_title", keep='last', inplace=True)
```

```python
# 4. Vérification finale
print("Aperçu des données :")
print(data.head())
```

```python
# 5. Sauvegarde finale avant l'ajout des films récents (2017-2026)
data.to_csv('main_data.csv', index=False)
print("Étape 2 terminée : main_data.csv prêt pour l'ajout des films Wikipédia.")
```
