```python
import pandas as pd
import numpy as np
```

```python
# 1. Chargement des données IMDB
data = pd.read_csv('movie_metadata.csv')
```

```python
# 2. On ne garde que les colonnes qui nous intéressent au début
data = data.loc[:,['director_name','actor_1_name','actor_2_name','actor_3_name','genres','movie_title']]
```

```python
# 3. Remplacement des valeurs manquantes (NaN) par 'unknown'
data['director_name'] = data['director_name'].replace(np.nan, 'unknown')
data['actor_1_name'] = data['actor_1_name'].replace(np.nan, 'unknown')
data['actor_2_name'] = data['actor_2_name'].replace(np.nan, 'unknown')
data['actor_3_name'] = data['actor_3_name'].replace(np.nan, 'unknown')
```

```python
# 4. Nettoyage de la colonne 'genres' (on remplace les | par des espaces)
data['genres'] = data['genres'].str.replace('|', ' ')
```

```python
# 5. Nettoyage des titres de films
data['movie_title'] = data['movie_title'].str.lower()
```

```python
# Supprimer le caractère spécial \xa0 à la fin des titres (très important !)
data['movie_title'] = data['movie_title'].apply(lambda x : x[:-1] if x.endswith('\xa0') else x)
```

```python
# 6. Sauvegarde pour l'étape suivante
data.to_csv('data_step1.csv', index=False)
print("Étape 1 terminée : data_step1.csv créé.")
```
